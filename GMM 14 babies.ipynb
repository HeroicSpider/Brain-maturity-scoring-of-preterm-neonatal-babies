```python
import numpy as np
import numpy as np
import pylab as py
import scipy as sc
import matplotlib.pylab as plt 
from sklearn.decomposition import FastICA
data = np.load('data.npy',allow_pickle=True)
import seaborn as sns
from sklearn import mixture
```

```python
X=np.c_[data[0][0],data[0][1],data[0][2]]
X1=X
len(X)
```

```python
gmm = mixture.GaussianMixture(n_components=2,covariance_type='full')

fig=plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
labels = gmm.fit_predict(X1)
ax.scatter(X1[:, 0],X1[:, 1],X1[:, 2], c=labels)
plt.title("GMM clustering")
plt.show()
print(gmm)
```

```python
plt.figure(figsize=(7,7))
loglike = gmm.score_samples(X1)
sns.histplot(loglike,kde=True)
plt.show()
```

```python
thres_LL = -100
ind_lowLL = np.where(loglike < thres_LL)
ind_lowLL
```

```python
t=np.linspace(0,2225275,2225275)
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(311)
ax = plt.plot(t,X1[:,0],color='red')
ax = plt.stem(t[ind_lowLL],X1[ind_lowLL,0].squeeze())
ax1 = fig.add_subplot(312)
ax1 = plt.plot(t,X1[:,1],color='red')
ax1=plt.stem(t[ind_lowLL],X1[ind_lowLL,1].squeeze())
ax2 = fig.add_subplot(313)
ax2 = plt.plot(t,X1[:,2],color='red')
ax2=plt.stem(t[ind_lowLL],X1[ind_lowLL,2].squeeze())
plt.show()
```

```python
for i in range(0,14):
    X=np.c_[data[i][0],data[i][1],data[i][2]]
    gmm = mixture.GaussianMixture(n_components=2,covariance_type='full')
    fig=plt.figure(figsize=(7,7))
    ax = fig.add_subplot(projection='3d')
    labels = gmm.fit_predict(X)
    ax.scatter(X[:, 0],X[:, 1],X[:, 2], c=labels)
    plt.title("GMM clustering")
    plt.show()
    print(gmm)
    plt.figure(figsize=(7,7))
    loglike = gmm.score_samples(X)
    sns.histplot(loglike,kde=True)
    plt.show()
```
